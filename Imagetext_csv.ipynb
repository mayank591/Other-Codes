```python
from PIL import Image
from pytesseract import pytesseract
import glob
import re
import os
```

```python
image_path='C:/Mayank/Projects/Project_APDCL/FieldData/*.jpg'
```

```python
image_list=glob.glob(image_path)
```

```python
pytesseract.tesseract_cmd='C:/Program Files/Tesseract-OCR/tesseract.exe'
```

```python
imagename=[]
lat=[]
long=[]
```

```python
for images in image_list:
    imagename.append(os.path.basename(images))
    img=Image.open(images)
    text=pytesseract.image_to_string(img)
    completetext=re.split(" |\n",text)
    if "Lat" in completetext:
        latindex=completetext.index('Lat')
        lat.append(completetext[latindex+1])
    else:
        lat.append("")
    if "Long" in completetext:
        lonindex=completetext.index('Long')
        long.append(completetext[lonindex+1])
    else:
        long.append("")
print(imagename)
print(lat)
print(long)
```

```python
print(len(imagename),len(lat),len(long))
```

```python
import pandas
df = pandas.DataFrame(data={"ImageName": imagename, "Lat": lat, "Long":long})
```

```python
df.to_csv("C:/Mayank/Projects/Project_APDCL/FieldData/Coordinates_Images.csv", sep=',',index=False)
```
